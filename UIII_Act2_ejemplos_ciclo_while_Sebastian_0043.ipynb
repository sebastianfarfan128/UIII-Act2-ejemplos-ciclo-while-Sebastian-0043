#### Ejemplo de ciclos while

```python
# Variable que inicia en 1
contador = 1

# Mientras el contador sea menor o igual a 5
while contador <= 5:

    # Mostrar el valor del contador
    print("Número:", contador)

    # Aumentar el contador en 1
    contador += 1
```

```python
# Variable vacía para guardar la contraseña
contraseña = ""

# Mientras la contraseña sea diferente
while contraseña != "python123":

    # Pedir la contraseña al usuario
    contraseña = input("Ingresa la contraseña: ")

# Mensaje cuando la contraseña es correcta
print("Acceso permitido")
```

```python
# Variable para guardar la suma
suma = 0

# Pedir el primer número
numero = int(input("Ingresa un número (0 para terminar): "))

# Mientras el número sea diferente de 0
while numero != 0:

    # Sumar el número a la variable suma
    suma += numero

    # Pedir otro número
    numero = int(input("Ingresa otro número: "))

# Mostrar el resultado final
print("La suma total es:", suma)
```

```python
# Variable para guardar la opción
opcion = ""

# Mientras la opción no sea 3
while opcion != "3":

    # Mostrar el menú
    print("\nMENÚ")
    print("1. Saludar")
    print("2. Despedirse")
    print("3. Salir")

    # Pedir una opción al usuario
    opcion = input("Elige una opción: ")

    # Si elige 1
    if opcion == "1":
        print("Hola")

    # Si elige 2
    elif opcion == "2":
        print("Adiós")

    # Si elige 3
    elif opcion == "3":
        print("Programa terminado")

    # Si escribe otra cosa
    else:
        print("Opción no válida")
```

```python
# La edad inicia en un número negativo
edad = -1

# Mientras la edad sea menor que 0
while edad < 0:

    # Pedir la edad
    edad = int(input("Ingresa tu edad: "))

    # Verificar si sigue siendo negativa
    if edad < 0:
        print("La edad no puede ser negativa")

# Mostrar la edad correcta
print("Tu edad es:", edad)
```

```python
# Número secreto
numero_secreto = 7

# Variable para guardar el intento
intento = 0

# Mientras el intento sea diferente
while intento != numero_secreto:

    # Pedir un número al usuario
    intento = int(input("Adivina el número: "))

    # Verificar si es incorrecto
    if intento != numero_secreto:
        print("Incorrecto, intenta otra vez")

# Mensaje cuando adivina
print("¡Correcto!")
```

```python
# Variable para guardar el texto
texto = ""

# Mientras el usuario no escriba salir
while texto != "salir":

    # Pedir una palabra
    texto = input("Escribe una palabra: ")

    # Variable para contar vocales
    vocales = 0

    # Recorrer cada letra
    for letra in texto:

        # Verificar si es vocal
        if letra in "aeiou":
            vocales += 1

    # Mostrar cuántas vocales tiene
    print("Tiene", vocales, "vocales")
```
